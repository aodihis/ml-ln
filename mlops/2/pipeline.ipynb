## Import

```python
import os
from sys import modules

from tfx.orchestration.beam.beam_dag_runner import BeamDagRunner
from modules.components import init_components, PipelineConfig
from modules.pipelines import init_pipeline
```

## Set variable

```python
PIPELINE_NAME = "aodihis-pipeline"

# pipeline inputs
DATA_ROOT = "data"
TRANSFORM_MODULE_FILE = "modules/transform.py"
TUNER_MODULE_FILE = "modules/tuner.py"
TRAINER_MODULE_FILE = "modules/trainer.py"
serving_model_dir = 'serving_model'
pipeline_root = PIPELINE_NAME
metadata_path = os.path.join(pipeline_root, "metadata.sqlite")
```

## Run pipeline

```python
config = PipelineConfig(
    DATA_ROOT,
    training_module=TRAINER_MODULE_FILE,
    transform_module=TRANSFORM_MODULE_FILE,
    tuner_module=TUNER_MODULE_FILE,
    training_steps=5000,
    eval_steps=1000,
    serving_model_dir=serving_model_dir,
)
components = init_components(
    config
)

pipeline = init_pipeline(components, pipeline_root, PIPELINE_NAME, metadata_path)
BeamDagRunner().run(pipeline=pipeline)
```
